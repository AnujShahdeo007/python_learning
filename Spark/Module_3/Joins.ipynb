Joins

A join combines data from two DataFrames based on a condition.

df1.join(df2, "column_name", "join_type")



```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.window import Window
from pyspark.sql.types import *

spark = (
    SparkSession.builder
    .appName("Joins")
    .master("local[*]")   
    .getOrCreate()
)

spark
```

```python
emp = [
    (1, "Alice", 10),
    (2, "Bob", 20),
    (3, "Charlie", 30),
    (4, "David", None)
]

emp_df = spark.createDataFrame(emp, ["emp_id", "name", "dept_id"])
```

```python
dept = [
    (10, "HR"),
    (20, "IT"),
    (40, "Finance")
]

dept_df = spark.createDataFrame(dept, ["dept_id", "dept_name"])
```

```python
emp_df.show()
dept_df.show()
```

Join Type       Use Case
Inner           Matching records
Left            Keep all left data
Right           Keep all right data
Outer           Combine everything
Left Anti       Find missing records
Left Semi       Filter matching rows




```python
#1. INNER JOIN 
#Only matching records from both tables
emp_df.join(dept_df, "dept_id", "inner").show()


```

```python
#LEFT JOIN 
#Take all records from LEFT table (emp)
#Match from right if available

emp_df.join(dept_df, "dept_id", "left").show()
```

```python
#3. RIGHT JOIN
#Take all records from RIGHT table

emp_df.join(dept_df, "dept_id", "right").show()
```

```python
#FULL OUTER JOIN
#Take everything from both tables
emp_df.join(dept_df, "dept_id", "outer").show()

```

```python
#LEFT ANTI JOIN
#Return rows in LEFT that DO NOT MATCH
emp_df.join(dept_df, "dept_id", "left_anti").show()

```

```python
#LEFT SEMI JOIN
#Return rows in LEFT that MATCH (but no right columns)

emp_df.join(dept_df, "dept_id", "left_semi").show()
```

Level 1 — Basic join understanding

1. Inner join

You have:
	•	employees(emp_id, emp_name, dept_id)
	•	departments(dept_id, dept_name)

Write a PySpark query to get all employees with their department names, but only where department exists in both tables.

⸻

2. Left join

Using the same tables, write a query to get all employees and their department names, even if some employees do not belong to any department.

⸻

3. Right join

Write a query to show all departments and the employees assigned to them, including departments with no employees.

⸻

4. Full outer join

Write a query to combine employees and departments so that:
	•	all employees appear
	•	all departments appear
	•	unmatched rows from either side are also included

⸻

5. Left anti join

Find all employees whose dept_id does not exist in the departments table.

⸻

Level 2 — Semi, anti, filtering-based joins

6. Left semi join

Find all employees who belong to a valid department, but return only employee columns.

⸻

9. Customers without orders

⸻

10. Employees with null department after left join
