```python
from pyspark.sql import SparkSession
spark=(
    SparkSession.builder
    .appName("LocalMachineExample")
    .master("local[*]") # local mode
    .getOrCreate()
)
# df=spark.range(1,11)
# df.show()
```

```python

spark.stop()
```

```python
# spark
sc=spark.sparkContext


print("SparkSession",spark)
print("SparkContext",sc.appName)
print("Master:",sc.master)
```

```python
data=list(range(1,21))
print(data)
```

```python
rdd=sc.parallelize(data,numSlices=4) #force 4 partition 
```

```python
rdd.getNumPartitions()
```

```python
print(rdd.glom().collect())
```

```python
mapped=rdd.map(lambda x:x*10)
```

```python
print(mapped.glom().collect())
```

```python
rdd=sc.parallelize([1,2,3,4,5,6,7,8],8)
print("Partitions before transformation")
print(rdd.glom().collect())

transformed=rdd.map(lambda x:x*10)
print("partitions after transformation (still lazy)")

print(transformed.glom().collect())
```

```python
pairs=sc.parallelize([
    ("a",1),("b",1),
    ("a",1),("b",1)
],2)

print("before")

print(pairs.glom().collect())

reduce=pairs.reduceByKey(lambda a,b:a+b)

print("after ReduceByKey partitions")

print(reduce.glom().collect())
```

```python
rdd=sc.parallelize([1,2,3],1)
```

```python
rdd.map(lambda x:[x,x*10]).collect() # Notice Output Nested List if we will use map()
```

```python
rdd.flatMap(lambda x:[x,x*10]).collect()
```

```python
rdd_2=sc.parallelize([1,2,3,4,5,6],2)
rdd_2.glom().collect()
```

```python
rdd_2.map(lambda x:x*2).collect()
```

# yield

```python
def numbers():
    return [1,2,3,4,5]

print(numbers())
```

```python
def numbers():
    yield 1
    yield 2
    yield 3
    yield 4
    yield 5

print(numbers())
```

```python
for n in numbers():
    print(n)
```
