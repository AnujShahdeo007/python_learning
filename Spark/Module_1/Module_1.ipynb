```python
rdd=sc.parallelize([1,2,3,4,5,6,7,8],8)
print("Partitions before transformation")
print(rdd.glom().collect())

transformed=rdd.map(lambda x:x*10)
print("partitions after transformation (still lazy)")

print(transformed.glom().collect())
```

```python
pairs=sc.parallelize([
    ("a",1),("b",1),
    ("a",1),("b",1)
],2)

print("before")

print(pairs.glom().collect())

reduce=pairs.reduceByKey(lambda a,b:a+b)

print("after ReduceByKey partitions")

print(reduce.glom().collect())
```

```python
# Joins 
emp=sc.parallelize([
    (1,"A"),
    (2,"B"),
    (3,"C"),
    (4,"D")
])

salary=sc.parallelize([
    (1,50000),
    (2,60000),
    (3,70000)
])
```

```python
result=emp.join(salary)
```

```python
result.collect()
```

```python
result_1=emp.leftOuterJoin(salary)
result_1.collect()
```

```python
result_3=emp.cogroup(salary)
```

```python
result_3.collect()
```

```python
# (key,(value from rdd1,value from rdd2))
for key,value in result_3.collect():
    print(key,list(value[0]),list(value[1]))
```

```python
for key,value in result_3.collect():
    print(key,value[0])
```

```python
result_3.mapValues(lambda x:(list(x[0]),list(x[1]))).collect()
```
