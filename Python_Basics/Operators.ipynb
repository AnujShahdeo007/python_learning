What is an Operator?

Definition

An operator is a symbol or keyword that tells Python to perform an operation on one or more values (operands).

Types of Operators in Python

Python has 7 major categories:
	1.	Arithmetic Operators
	2.	Assignment Operators
	3.	Comparison (Relational) Operators
	4.	Logical Operators
	5.	Bitwise Operators
	6.	Membership Operators
	7.	Identity Operators

#Membership Operators
Operator                 Meaning
in                       Evaluates to true if it finds a variable in the specified sequence and false otherwise.
not in                   Evaluates to true if it does not find a variable in the specified sequence and false otherwise.

value in collection 
value not in collection

Membership Operators Supported Datatypes 
str
list 
tuple 
set 
dict( check keys only but not values)
custom objects (via __contains__ )

```python
text="Python programming"
print("Python" in text)
print("Java" in text)
print("prog" in text)
```

```python
word="dataengineering"
if "engine" in word:
    print("Found")
```

```python
numbers=[10,20,30,40,50]
print(25 in numbers)
print(30 in numbers)
```

```python
users=["admin","user1","user2","user3"]
username="user3"
if username not in users:
    print("Access Denied")
else:
    print("Access Granted")
```

```python
coords=(10,20,30)
print(10 in coords)
```

```python
ids={1001,1002,1003}
print(1004 not in ids)
print(1002 in ids)
print(1005 in ids)
```

```python
lst=list(range(1000000))
st=set(lst)
print(999999 in lst) # Slower for large datasets
print(999999 in st) # Fastest way to check membership in large datasets
```

```python
student={
    "Name":"AWS",
    "age":25
}
print("Name" in student)
print("grade" not in student)   
print("AWS" in student)
```

```python
print("AWS" in student.values())
```
