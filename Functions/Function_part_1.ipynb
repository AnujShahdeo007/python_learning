```python
def calculate_discount(price, discount_percentage):
    discount= (price * discount_percentage)/100
    final_price= price - discount
    print("The final price after discount is:", final_price)
    return final_price
```

```python
abc1=calculate_discount(2000, 50)
```

```python
abc1
```

```python
total_price=abc1+ 100  # adding shipping cost
print("Total price including shipping is:", total_price)
```

```python
import time 
start_time= time.time()
total=0
for i in range(1,1000001):
    total += i
end_time= time.time()
print("Sum is:", total)
print("Time taken to compute the sum:", end_time - start_time, "seconds")
```

```python
import time
def calculate_sum():
    total=0
    for i in range(1,1000001):
        total += i
    return total
start_time= time.time()
sum= calculate_sum()
end_time= time.time()
print("Sum is:", sum)
print("Time taken to compute the sum:", end_time - start_time, "seconds")   
```

```python
def even_odd(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"
```

```python
even_odd(5)
```

```python
even_odd=lambda num:"Even" if num%2==0 else "Odd"
```

```python
print(even_odd(5))
```

```python
def countdown(n):
    if n==0:
        return
    print(n)
    countdown(n-1) 
```

```python
countdown(0)
```

```python
def factorial(n):
    if n==0 or n==1:
        return 1
    return n*factorial(n-1)
```

```python
factorial(4)
```
