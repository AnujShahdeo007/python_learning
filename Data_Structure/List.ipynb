Shallow Copy :

Photocopy of a file that still points to the same attched documents

Deep Copy :

Completly new file with all documents duplicated 

Feature             Shallow copy                Deep Copy 

Outer Object        New                             New 
Inner Object        Shared                          New 
Speed               Faster                          Slower 
Nested              No                              Yes
Memory Usages       Less                            More 


When Deep copy is Required?
- Nested List 
- Dict if list 
- Json like data 
- Data pipeline ( Spark )


2. Why shallow copy is dangerous?

----> Because modifiting nested objects affects original 


3. Is slicing deep Copy?

----> No, Slicing is shallow copy 


4. Why deepcopy is slower ?

----> Because it recursively copies everything.
