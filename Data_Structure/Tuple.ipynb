Use List when Data is Dynamic 
    - Add/remove/update 
    - Size of data is Unknown 
    - Data changes frequently 
    
    API Responses 

Use Tuple when data is FIXED/CONSTANT 
    - Data structure should never change 
    - Fixed number of values 
    - Represents single records 
    - Return multiple values from function
