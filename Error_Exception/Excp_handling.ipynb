```python
try:
    num=int(input("Enter a number: "))
    print(100/num)
except ValueError:
    print("Invalid input. Please enter a valid number.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
```

```python
try:
    n=int(input("Enter number:"))
except ValueError:
    print("Invalid Input")
else:
    print("Square is", n*n)

# Working 
    #- If exception -> except run, else skkiped 
    #- if no exception -> Else runs  
```

```python
try:
    f=open("data.txt","r")
    print(f.read())
except FileNotFoundError:
    print("File not found")
finally:
    print("Execution completed")

# Wheather file exists or not, finally will run. It is used to release resources like closing file, database connection etc.
```
