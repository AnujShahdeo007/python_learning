2.	Write a program to check whether a number is a strong number (sum of factorials of digits = number).

- Extract each digit from the number 
- Find the factorial of each digit 
- Keep adding factorials 
- Compare sum with original number 



```python
num=int(input("Enter a number:"))
temp=num
sum_fact=0

while temp>0:
    digit=temp%10
    fact=1
    for i in range(1,digit+1):
        fact=fact*i
    sum_fact=sum_fact+fact
    temp=temp//10
if sum_fact==num:
    print(f"{num} is a strong number")
else:
    print(f"{num} is not a strong number")
```
